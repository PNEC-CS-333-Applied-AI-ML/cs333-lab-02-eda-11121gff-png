## 4ï¸âƒ£ Outlier Detection

```python

# Boxplot, IQR, Z-modified score
df.boxplot(column=['Engine HP'])
```

```python
df.boxplot(column=['MSRP'])
```

```python
df.boxplot(column=['Engine Cylinders'])
```

```python
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

```python
# Calculate lower and upper bounds for outlier detection using IQR method
lower_bounds = Q1 - 1.5 * IQR
upper_bounds = Q3 + 1.5 * IQR

# Create a boolean mask to identify outliers only within numerical columns
# We use the index of Q1 (which contains the names of numerical columns) to select relevant columns from df
outlier_mask = ((df[lower_bounds.index] < lower_bounds) | (df[upper_bounds.index] > upper_bounds))

# Filter the DataFrame: keep rows where NO numerical column is an outlier
df = df[~outlier_mask.any(axis=1)]
df.shape
```


### âœï¸
- Which method did you use (IQR or Modified Z-score)?

  ANSWR: IQR
- Why?

  ANSWER: Because it looks at the middle 50% of the data and identifies values far from this central range. It is robust to skewed data and extreme values, unlike mean-based methods that can be distorted by outliers. Since your dataset has features like MSRP and Engine HP with very high values, IQR reliably detects true outliers without being misled by them.

- How many outliers were removed?

  ANSWER: 2

- How did it affect statistics?

  ANSWER: It will make the data more stable. The mean, variance, standard deviation, and range will decrease, moving closer to typical values, while the median and IQR will stay mostly the same. Overall, the distribution becomes less skewed and more representative of the majority of the data.
